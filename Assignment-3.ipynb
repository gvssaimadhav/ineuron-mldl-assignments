## Q 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()


```python
def myreduce(func, my_ls):
    result = my_ls[0]
    for i in my_ls[1:]:
        result = func(result, i)
    return result


#testing myreduce()
def sum(x,y): 
    return x + y

print (myreduce(sum,[5, 10, 15, 20]))
```

## Q 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def myfilter(anyfunc, sequence):
    result = []# iterate over sequence of items in sequence and apply filter function
    for i in sequence:
        if anyfunc(i):
            result.append(i)
    return result

# test myfilter function
def ispositive(x):
    if (x <= 0): 
        return False 
    else: 
        return True
    
print(myfilter(ispositive,[-10,-9,-8, 8, 9, 10]))
```
